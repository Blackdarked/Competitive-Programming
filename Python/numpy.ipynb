```python
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)
pl.plot(x, y)
```

```python
dydx = np.gradient(y, x)
pl.plot(x, dydx)
```

```python
y_int = np.cumsum(y) * (x[1] - x[0 ])
pl.plot(x, y_int)
```
